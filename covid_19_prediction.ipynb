```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
covid_df=pd.read_csv('/content/Cleaned-Data.csv')
```

```python
covid_df.head()
```

```python
print(covid_df.head())
```

```python
# Separate features (x) and target (y)
x = covid_df.drop(['Severity_Severe'], axis=1)
y = covid_df['Severity_Severe']

# Convert categorical 'Country' column to numerical using one-hot encoding
x = pd.get_dummies(x, columns=['Country'], drop_first=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model_selection = RandomForestRegressor(n_estimators=100, random_state=42)
model_selection.fit(X_train, y_train)
```

```python
y_prediction=model_selection.predict(X_test)
```

```python
accuracy_of_model=accuracy_score(y_test,y_prediction)
```

```python
print("ACCURACY:",accuracy_of_model)
```

```python
cm = confusion_matrix(y_test, y_prediction)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
